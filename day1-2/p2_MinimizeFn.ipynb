```python
# Find the minimum value of the function f(x) = x**4 - 3*(x) - 2

x = tf.Variable(0.45, dtype = tf.float32)
f = tf.pow(x, 4) - 3*x - 2

loss = f
#gd = tf.train.GradientDescentOptimizer(learning_rate = 0.001) #parameters learning rate
gd = tf.train.MomentumOptimizer(momentum = 0.01, learning_rate = 0.001)
opt = gd.minimize(loss)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(100):
        print(sess.run([x, loss]))
        sess.run(opt)
```
