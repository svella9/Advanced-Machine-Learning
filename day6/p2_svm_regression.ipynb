```python
import csv
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
```

```python
x = np.sort(5 * np.random.rand(40, 1), axis= 0)
y = np.sin(x).ravel()
```

```python
clf = svm.SVR(kernel = 'linear', epsilon = 0.2, C = 2)
y_pred = clf.fit(x,y).predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, y_pred, color = 'blue')
print('Linear')
plt.show()
```

```python
clf = svm.SVR(kernel = 'rbf', epsilon = 0.2, C = 2)
y_pred = clf.fit(x,y).predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, y_pred, color = 'blue')
print('Gaussian')
plt.show()
```

```python
clf = svm.SVR(kernel = 'poly', degree = 3, epsilon = 0.2, C = 2)
y_pred = clf.fit(x,y).predict(x)
plt.scatter(x, y, color = 'red')
plt.plot(x, y_pred, color = 'blue')
print('Polynomial')
plt.show()
```
